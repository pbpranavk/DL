## Create two 1D Tensors of same length in PyTorch with arbitrary values. How do you determine if the angle between the two tensors is < 90 degrees or > 90 degress? Demonstrate this using code.

```python
x = torch.tensor([1, 4])
y = torch.tensor([-1, -4])
```

```python
sum_val = sum(x * y)
```

```python
sum_val
```

```python
mag_x = torch.sqrt(torch.sum(torch.square(x)))
mag_y = torch.sqrt(torch.sum(torch.square(y)))
```

```python
total_mag = mag_x * mag_y
```

```python
total_mag
```

```python
total_inc = sum_val / total_mag
```

```python
total_inc
```

```python
radians = torch.acos(total_inc)
```

```python
theta = radians * 57.295
```

```python
theta
```

## As we can see the angle between these two vectors in 180 degrees

## These two are exactly opposite to each other

# The way to find the angle between these two vectors is the inverse cos of the dot product over the product of magnitudes of the two vectors

# The value we get is in radians we need to multiply it with 57.295 to get the value in degrees by this we can find if the value is less than 90 or not
