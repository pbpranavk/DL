## Solution for 1d

```python
def calc_least_square(a, b, list1):
  sum_of_squares = 0
  for [xi, yi] in list1:
    sum_of_squares += (yi - (a * xi + b)) ** 2
  return sum_of_squares / (2 * len(list1))
```

```python
val_0_0 = calc_least_square(0, 0, [[2, 1], [3,2], [4,3]] )
```

```python
val_0_1 = calc_least_square(0, 1, [[2, 1], [3,2], [4,3]] )
```

```python
val_0_2 = calc_least_square(0, 2, [[2, 1], [3,2], [4,3]] )
```

```python
val_1_0 =calc_least_square(1, 0, [[2, 1], [3,2], [4,3]] )
```

```python
val_1_1 = calc_least_square(1, 1, [[2, 1], [3,2], [4,3]] )
```

```python
val_1_2 =calc_least_square(1, 2, [[2, 1], [3,2], [4,3]] )
```

```python
val_2_0 = calc_least_square(2, 0, [[2, 1], [3,2], [4,3]] )
```

```python
val_2_1 =calc_least_square(2, 1, [[2, 1], [3,2], [4,3]] )
```
